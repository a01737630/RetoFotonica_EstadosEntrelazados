```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
alfa = np.linspace(0, 2 * np.pi, 2**10)
beta = np.linspace(0, 2 * np.pi, 2**10)
ALFA, BETA = np.meshgrid(alfa, beta)

C = np.sin(ALFA-BETA)
p = 0.5*(1 + np.cos(ALFA-BETA))
q = 1 - p

S = -(np.where(p > 0, p*np.log2(p), 0.0) + np.where(q > 0, q*np.log2(q), 0.0))

fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(1, 2, 1,)
im1 = ax1.imshow(C, extent=(0, 2*np.pi, 0, 2*np.pi), origin='lower', cmap='viridis')
ax1.set_title('Concurrencia C')
ax1.set_xlabel('Alfa')
ax1.set_ylabel('Beta')
cbar1 = fig.colorbar(im1, ax=ax1, fraction=0.046, pad=0.04)

ax2 = fig.add_subplot(1, 2, 2)
im2 = ax2.imshow(S, extent=(0, 2*np.pi, 0, 2*np.pi), origin='lower', cmap='inferno')
ax2.set_title('Entropa de entrelazamiento S')
ax2.set_xlabel('Alfa')
ax2.set_ylabel('Beta')
cbar2 = fig.colorbar(im2, ax=ax2, fraction=0.046, pad=0.04)
plt.show()
```
